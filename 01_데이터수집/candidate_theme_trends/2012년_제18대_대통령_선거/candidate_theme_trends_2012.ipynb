```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

### 박근혜 후보

```python
# 박근혜 후보 테마주

pytrends = TrendReq(hl='ko', tz=540)    # pytrends 객체 생성 (한국 기준)
keywords = ["대유신소재", "대유에이텍", "EG", "보령메디앙스", "아가방컴퍼니"] # 키워드 생성, 최대 5개

# 데이터 요청
pytrends.build_payload(
    keywords,
    cat=0,
    timeframe='2012-04-23 2012-12-27',  # 날짜 선정 기준: 예비 후보자 등록 ~ 당선 확정 + 1주일
    geo='KR')

df = pytrends.interest_over_time()  # 관심도 데이터 추출

# 시각화
df.drop(columns='isPartial').plot(figsize=(12,6), title='2012 박근혜 후보 테마주 관심도 추이')
plt.xlabel('날짜')
plt.ylabel('관심도')
plt.grid()
plt.show()

# 저장하기
# df = df.drop(columns='isPartial')  # isPartial 컬럼 제거
# df.to_csv("../../data/2012_parkgeunhye_theme_trends.csv", encoding='utf-8-sig')
```

```python
grouped = dict(tuple(df.groupby([df.index.year, df.index.month])))  # 월별로 분리

# 시각화
for (year, month), data in sorted(grouped.items()):
    plt.figure(figsize=(12,6))
    data.plot(ax=plt.gca())
    plt.title(f'{year}년 {month}월 박근혜 후보 테마주 관심도')
    plt.xlabel('날짜')
    plt.ylabel('관심도')
    plt.grid()
    plt.tight_layout()
    plt.show()
```
